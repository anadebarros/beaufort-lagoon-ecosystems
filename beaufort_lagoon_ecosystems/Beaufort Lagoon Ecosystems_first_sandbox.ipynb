```python
import pandas as pd

df = pd.read_csv('BLE_LTER_sea_ice_measurements.csv', parse_dates=["date_time"]) #transforming date_time colum into a Datetime

#get name of columns into a list, makes it easier to see which columns there are and use their names
df.columns.values.tolist()
```

```python
#get only the columns we want
df1 = pd.DataFrame(df, columns=['lagoon','station_name','date_time','hole1_ice_thickness_m','hole2_ice_thickness_m',
                                'hole3_ice_thickness_m', 'spot1_snow_depth_m', 'spot2_snow_depth_m', 'spot3_snow_depth_m',
                                'spot4_snow_depth_m','spot5_snow_depth_m', 'spot6_snow_depth_m', 'spot7_snow_depth_m','spot8_snow_depth_m',
                                'spot9_snow_depth_m','spot10_snow_depth_m', 'freeboard_m', 'latitude', 'longitude'])

df1
```

```python
#group the data by lagoon name and get associated data
#use functions for each lagoon
def simpson_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Simpson')]

def stefansson_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Stefansson')]

def jago_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Jago')]

def kaktovik_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Kaktovik')]

def elson_east_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Elson East')]

def elson_west_func():
    return df1.iloc[df1.groupby("lagoon").indices.get('Elson West')]
```

```python
#itirate over the dataframe and create functions that return lists of the different columns I want
def date(data):
    date_list = []
    for i in range(len(data)):
        date_list.append(data.iloc[i, 2])
    return date_list
        
def h1(data):
    h1_list = []
    for i in range(len(data)):
        h1_list.append(data.iloc[i, 3])     
    return h1_list

def h2(data):
    h2_list = []
    for i in range(len(data)):
        h2_list.append(data.iloc[i, 4])
    return h2_list
    
#h3 does not have values for simpson
```

```python
from matplotlib import pyplot as plt

#plot data from h1 and h2. Used a plot for h1 values, and a scatter for h2
plt.plot(date(simpson_func()), h1(simpson_func()), 'rx', label='hole 1')
plt.scatter(date(simpson_func()), h2(simpson_func()))

plt.xlabel("Time (Y/M))")
plt.ylabel("Ice thickness (m)")

#define limits in y axis
plt.ylim(0, 3)

plt.show()
```
